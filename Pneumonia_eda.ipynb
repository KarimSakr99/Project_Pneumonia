```python
import numpy as np
import pandas as pd
from itertools import chain
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./Data_Entry_2017_v2020.csv')
df.head(2)
```

```python
labels=sorted(set(chain(*df['Finding Labels'].apply(lambda x:x.split('|')))))
```

```python
for label in labels:
    df[label]= df['Finding Labels'].apply(lambda x:1 if label in x else 0)
```

```python
df[['Image Index','Finding Labels']+labels]
```

```python
df['Pneumonia'].value_counts()
```

```python
df_positive=df[df['Pneumonia']==1]
df_positive
```

```python
df[df['Pneumonia']==0]
```

```python
pat_id_pneumonia=df[df['Pneumonia']==1]['Patient ID'].unique()
pat_id_pneumonia,pat_id_pneumonia.size
```

```python
df['had_pneumonia']=df['Patient ID'].apply(lambda x:1 if x in pat_id_pneumonia else 0)
```

```python
df[df['had_pneumonia']==1]['Patient ID'].nunique()
```
